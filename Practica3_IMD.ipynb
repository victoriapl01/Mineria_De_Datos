```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv('https://raw.githubusercontent.com/victoriapl01/Mineria_De_Datos/main/Datasets/store_data.csv')
print(dataset.shape)

# Transforming the list into a list of lists, so that each transaction can be indexed easier
transactions = []
for i in range(0, dataset.shape[0]):
    transactions.append([str(dataset.values[i, j]) for j in range(0, dataset.shape[1])])


from apyori import apriori
# Please download this as a custom package --> type "apyori"
# To load custom packages, do not refresh the page. Instead, click on the reset button on the Console.

rules = apriori(transactions, min_support = 0.005, min_confidence = 0.25, min_lift = 3, min_length = 2)
# Support: number of transactions containing set of times / total number of transactions
# .      --> products that are bought at least 3 times a day --> 21 / 7501 = 0.0027
# Confidence: Should not be too high, as then this wil lead to obvious rules

#Try many combinations of values to experiment with the model

results = []
for item in rules:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")

    results.append(item)


# viewing the rules
print(len(results))

# Transferring the list to a table
results = pd.DataFrame(results)

print(results.head(5))
```

Este código utiliza el algoritmo de aprendizaje automático de asociación Apriori para encontrar reglas de asociación en un conjunto de datos de transacciones de una tienda. El algoritmo se basa en la idea de que si un artículo es comprado con frecuencia en una transacción, también es probable que otros artículos se compren con ese artículo.

El código carga un conjunto de datos de un archivo CSV, luego utiliza el paquete "apyori" para aplicar el algoritmo Apriori al conjunto de datos de transacciones. Los parámetros utilizados son:

min_support: un valor flotante que especifica el nivel mínimo de soporte necesario para que una regla sea considerada. El soporte se refiere al número de transacciones que contienen un conjunto de artículos.
min_confidence: un valor flotante que especifica el nivel mínimo de confianza necesario para que una regla sea considerada. La confianza se refiere a la probabilidad de que un artículo dado sea comprado si otro artículo dado ya ha sido comprado.
min_lift: un valor flotante que especifica el nivel mínimo de "lift" necesario para que una regla sea considerada. El "lift" se refiere a la relación entre la probabilidad de comprar un conjunto de artículos juntos y la probabilidad de comprar cada artículo individualmente.
min_length: un entero que especifica el número mínimo de artículos en un conjunto para que una regla sea considerada.
Una vez que se han encontrado las reglas, el código imprime ciertos atributos de cada regla, como el soporte, la confianza y el "lift".

Es importante tener en cuenta que el valor de soporte, confianza y lift son considerados para determinar cuales de las reglas son importantes. Una regla con un alto soporte, confianza y lift es considerada como más importante. Sin embargo, también es importante tener en cuenta el contexto de los datos y el negocio para determinar qué reglas son realmente relevantes y útiles.





Esta es una lista de reglas de asociación que se han extraído de un conjunto de datos. Cada regla tiene tres medidas de rendimiento: soporte, confianza y elevación.

El soporte es la proporción de transacciones que contienen tanto el lado izquierdo (LHS) como el lado derecho (RHS) de la regla.

La confianza es la proporción de transacciones que contienen el LHS y también contienen el RHS. Mide con qué frecuencia el RHS está asociado con el LHS.

La elevación es la relación entre el soporte observado y el esperado si el LHS y el RHS fueran independientes. Mide con qué frecuencia se compra el RHS con el LHS en comparación con lo que se esperaría si el LHS y el RHS fueran independientes
